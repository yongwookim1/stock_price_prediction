```python
import requests
from bs4 import BeautifulSoup as bs
from datetime import datetime, timedelta
import pandas as pd
from tqdm import tqdm

# Set start data and end date
start_date = datetime.strptime('20230613', '%Y%m%d')
end_date = datetime.strptime('20240613', '%Y%m%d')

# Generate date list
date_list = [(start_date + timedelta(days=x)).strftime('%Y%m%d') for x in range((end_date - start_date).days + 1)]

news = []

# Crawling
for date in tqdm(date_list):
    url = f'https://news.daum.net/breakingnews/economic/stock?regDate={date}'
    res = requests.get(url)
    soup = bs(res.text, 'lxml')
    ul = soup.find("ul",{"class":"list_news2"})
    if ul:
        li_list = ul.findAll("li")
        for li in li_list:
            data = li.find("a", {"class": "link_txt"})
            if data:
                formatted_date = datetime.strptime(date, '%Y%m%d').strftime('%Y-%m-%d')
                news.append({
                    'Date': formatted_date,
                    'title': data.text.strip(),
                })
            break

# Generate dataframe
df = pd.DataFrame(news)

# Save csv file
df.to_csv('news_titles.csv', index=False, encoding='utf-8-sig')

print("Success!")
df.head()
```
